```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Membaca data dari file CSV
data = pd.read_csv('/content/Tesla.csv - Tesla.csv.csv')

# Mengubah kolom 'tanggal' menjadi tipe data datetime
data['Date'] = pd.to_datetime(data['Date'])

# Memisahkan data menjadi dua kolom: x (tanggal) dan y
x = data['Date']
y = data['Close']

# Membuat line chart
plt.plot(x, y)

# Menambahkan judul dan label sumbu
plt.title('Tesla Stock Prices')
plt.xlabel('Year')
plt.ylabel('Price ($)')

# Mengatur keterangan tanggal menjadi miring
plt.xticks(rotation=45)

# Menampilkan chart
plt.show()
```

Line chart menunjukkan harga saham Tesla yang telah mengalami peningkatan selama periode tujuh tahun. Pada tahun 2011, harga sahamnya sekitar 20 USD dan meningkat menjadi 250 USD pada tahun 2014.
